```python
import numpy as np
import matplotlib.pyplot as plt

# Inicializando a lista para armazenar os resultados das simulações
null_outcomes = []

# Simulando 10.000 experimentos
for i in range(10000):
    simulated_monthly_visitors = np.random.choice(['y', 'n'], size=500, p=[0.1, 0.9])
    num_purchased = np.sum(simulated_monthly_visitors == 'y')
    null_outcomes.append(num_purchased)

# Convertendo para array do numpy para facilitar operações
null_outcomes = np.array(null_outcomes)


# Cálculo dos Valores Mínimo e Máxim    o
null_min = np.min(null_outcomes)
null_max = np.max(null_outcomes)
print("Valor mínimo:", null_min)
print("Valor máximo:", null_max)

# Cálculo do valor p
p_value = np.sum((null_outcomes <= 41) | (null_outcomes >= 59)) / len(null_outcomes)
print("Valor p:", p_value)

# Plotando o histograma dos resultados das simulações
plt.hist(null_outcomes, bins=range(0, 101), color='blue', edgecolor='black', density=True, alpha=0.7, label='Distribuição das Compras')

# Adicionando linhas verticais para os valores extremos
plt.axvline(41, color='red', linestyle='dashed', linewidth=1, label='Limite Inferior (41)')
plt.axvline(59, color='green', linestyle='dashed', linewidth=1, label='Limite Superior (59)')

# Adicionando título e legendas
plt.title('Distribuição do Número de Compras Simuladas')
plt.xlabel('Número de Compras')
plt.ylabel('Frequência')
plt.legend()

# Exibindo o gráfico
plt.show()
```
